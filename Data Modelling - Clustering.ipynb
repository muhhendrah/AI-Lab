```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
```

```python
plt.style.use('seaborn')
```

### Genarate Data

```python
X,_ = make_blobs(n_samples= 800, n_features= 2,
                centers = 3, cluster_std = 1.9,
                random_state = 20, shuffle = True
                )
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1], c= 'white', edgecolor='r', marker = 'o', s=50)
plt.show()
```

## Model Data

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2, init = 'random', max_iter = 300, random_state=20)
y_km = km.fit_predict(X)
```

### Visualisasi Cluster

```python
plt.subplots(figsize=(10,7))
plt.scatter(
    X[y_km == 0,0], X[y_km == 0,1],
    c = 'green', edgecolor = 'black',
    label = 'cluster 1'
)
plt.scatter(
    X[y_km == 1,0], X[y_km == 1,1],
    c = 'blue', edgecolor = 'black',
    label = 'cluster 2'
)
#plt.scatter(
#    X[y_km == 2,0], X[y_km == 2,1],
#    c = 'red', edgecolor = 'black',
#    label = 'cluster 3'
#)

#plot centroid
plt.scatter(
    km.cluster_centers_[:,0], km.cluster_centers_[:,1],
    c = 'orange', edgecolor = 'black', label='cantroid'
)

plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(X)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels,metric='euclidean')
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(X)
    distortions.append(km.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```
